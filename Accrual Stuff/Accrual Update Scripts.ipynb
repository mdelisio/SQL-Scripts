Check Project using ProjectID in Fact table

```sql
Select *
FROM dw.ProjectGrossMarginAdjustmentActivityFact pgmaaf
INNER JOIN edw.ProjectDim pd
    ON pgmaaf.ProjectDimKey = pd.ProjectDimKey
WHERE pd.CompanyID = 'Solar'
AND pd.ProjectID = '2087293'
Order by BusinessDateDimKey Desc
```

Look for Latest Acconting Period End Date

```sql
SELECT TOP (1000) 
      [CalendarDate]  
      ,[AccountingPeriodName]
      ,[AccountingPeriodSequenceNumber]
  FROM [EDW].[DateDim]
  Where CalendarDate <= GETDATE()
  Order by CalendarDate DESC
```
