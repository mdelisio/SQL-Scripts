# **P&M Order Business Logic**

```sql
--5005652 Order Num is shipped in Pack 200938 
--5005660 is an Open order and has no Pack number
--2257001 is an Open Order and has 13 pack numbers. The OrderLines are all closed that are tied to these packs
--5005399 is a closed order that has 28 packs on 7 order line releases with different Ship Dates
--

--Declare @PackNum int
--Set @Packnum = 200938;
Declare @Company Varchar(50)
Set @Company = 'TES';
Declare @OrderNum int
Set @OrderNum = 5005399

/*
Select *
From erp.ShipHead
Where Company = @Company
and packnum = @Packnum
Order by ShipDate DESC
*/


Select 
SD.Company, SD.PackNum, SD.OrderNum, SD.Orderline, SD.OrderRelNum, SH.ShipDate, OL.OpenLine
FROM erp.ShipDtl SD
LEFT JOIN erp.ShipHead SH
    ON SD.Company = SH.Company
        AND SD.PackNum = SH.PackNum
LEFT JOIN erp.OrderDtl OL 
    ON SD.Company = OL.Company
        AND SD.OrderNum = OL.OrderNum
        AND SD.OrderLine = OL.OrderLine
Where SD.Company = @Company
and SD.Ordernum = @OrderNum

SELECT DISTINCT 
Company,
OrderNum,
Count(PackNum) PackCount
FROM erp.ShipDtl
Where Company = @Company
and OrderNum = @OrderNum
Group By Company, Ordernum


Select Top (10)  *
From erp.ShipDtl
Where Company = @Company
and Ordernum = @OrderNum

Select *
From erp.OrderHed
Where Company = @Company
and OrderNum = @OrderNum

Select *
From erp.Orderdtl
Where Company = @Company
and OrderNum = @OrderNum

Select *
From erp.OrderRel
Where Company = @Company
and OrderNum = @OrderNum

```

```sql
Select Top (20) *
From erp.OrderHed
Where Company = 'TES'
and OpenOrder = 0
And VoidOrder = 0
Order by Ordernum DESC

```

```sql
SELECT      COLUMN_NAME AS 'ColumnName'
            ,TABLE_NAME AS  'TableName'
            ,TABLE_SCHEMA as 'SchemaName'
FROM        INFORMATION_SCHEMA.COLUMNS
WHERE       COLUMN_NAME LIKE 'PackNum%'
and         Table_Schema = 'erp'
ORDER BY    ColumnName
```

```sql
WITH ShipOrderSummary AS
(SELECT
Company,
OrderNum,
Count(PackNum) as PackCount
FROM Erp.ShipOrd
Group By Company, OrderNum
)

Select 
Company,
OrderNum,
PackCount
FROM ShipOrderSummary
Where Company = 'RBMI'
and PackCount > 1
Order by PackCount DESC
```
