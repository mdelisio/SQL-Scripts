```sql
Select
 oh.OrderNum,
 c.Name,
 datepart(yyyy, od.ChangeDate) AS OrderYear,
 sum(od.UnitPrice) as TotalSellPrice,
 st.State,
 oh.Division_c

From [ERPDB].[erp].[OrderDtl] AS od
Left JOIN [ERPDB].dbo.[OrderHed] as oh
On od.OrderNum = oh.OrderNum
Left Join ERPDB.erp.Customer AS c
on oh.CustNum = c.CustNum
Left Join [ERPDB].[Erp].[ShipTo] st 
on oh.CustNum=st.CustNum and oh.ShipToNum=st.ShipToNum
Where od.Company = 'Solar' and od.ChangeDate >= '01/01/2022' and c.Company = 'Solar' and od.Unitprice > 0
Group By oh.OrderNum, c.Name, datepart(yyyy, od.ChangeDate), st.State, oh.Division_c
```
